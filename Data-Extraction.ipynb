```python
dataframes = []
counter = 1

for file in tqdm(filenames):
    with open(file, 'r') as f:
        try:
            # Load and normalize data
            content = safe_load(f)
            if content:  # Check if content is not empty
                df = pd.json_normalize(content)
                df['match_id'] = counter
                dataframes.append(df)  # Add each DataFrame to the list
                counter += 1
            else:
                print(f"Skipping empty file: {file}")
        except NotImplementedError:
            print(f"Skipping file {file} due to incompatible format.")
        except Exception as e:
            print(f"Error processing file {file}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if dataframes:
    final_df = pd.concat(dataframes, ignore_index=True)
else:
    final_df = pd.DataFrame()  # Create an empty DataFrame if no valid data

final_df
```

```python
count = 1
dataframes = []  # List to store individual DataFrames for each match

# Iterate over each match
for index, row in matches.iterrows():
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count += 1
        continue
    count += 1

    # Lists to store data for the current match
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []

    # Process deliveries in the 1st innings
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            
            # Handle player dismissal information
            player_of_dismissed.append(ball[key].get('wicket', {}).get('player_out', '0'))

    # Create a DataFrame for the current match and add it to the list
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    dataframes.append(loop_df)

# Concatenate all individual match DataFrames into a single DataFrame
delivery_df = pd.concat(dataframes, ignore_index=True)

```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
