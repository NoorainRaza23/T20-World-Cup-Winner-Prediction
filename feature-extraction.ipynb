Calculates the Cumulative Sum of Runs for Each Match

```python
df = df.copy()
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

Extracting the Over Number & Ball Number within the Over:

```python
df.loc[:, 'over'] = df['ball'].apply(lambda x: str(x).split(".")[0])
df.loc[:, 'ball_no'] = df['ball'].apply(lambda x: str(x).split(".")[1])
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
# Ensure that 'player_dismissed' is treated as numeric, handling possible non-numeric values
df['player_dismissed'] = pd.to_numeric(df['player_dismissed'], errors='coerce')  # This will convert non-numeric values to NaN
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)  # Convert 0s to 0, others to 1
df['player_dismissed'] = df['player_dismissed'].astype('int')  # Convert to integer type

# Now perform the cumulative sum
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()

# Calculate wickets left
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df
```

```python
# Convert 'runs' to numeric, coercing errors to NaN, then drop NaNs
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')
df = df.dropna(subset=['runs'])
```

```python
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []

for id in match_ids:
    group_runs = groups.get_group(id)['runs']
    last_five.extend(group_runs.rolling(window=30).sum().tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python

final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
import pickle
import pandas as pd

# Load the pickle file
final_df = pickle.load(open('dataset_level5.pkl', 'rb'))

# Convert the DataFrame to CSV
final_df.to_csv('dataset_level5.csv', index=False)

print("File converted to CSV and saved as 'dataset_level5.csv'")
```

```python
final_df = final_df.sample(final_df.shape[0])    #suffling the data unbaising
```

```python
final_df
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python

pickle.dump(pipe,open('pipe.pkl','wb'))
```
